####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
# lets create a seperate column for discounted_price and name it 'Revenue_List'
def Revenue_List(x):
    if x>600:
        return x*0.25
    elif x>350:
        return x*0.15
    elif x>100:
        return x*0.10
    else:
        return x*0.05
    
ECom_Data['Company_Revenue']= ECom_Data['discounted_price'].apply(Revenue_List)

ECom_Data.head()
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


### Ans 6

```python
ECom_Data['Brand_Revenue']= ECom_Data['discounted_price']-ECom_Data['Company_Revenue'] 
REV=ECom_Data.groupby('Brand').sum('Brand_Revenue')
REV1=REV['Brand_Revenue'].sort_values(ascending=False).head(10)
print(REV1)
ECom_Data.head()

```

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and 
## iv.	total BrandRevenue generated per Brand. 

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

```python
Orders_per_Brand= ECom_Data.groupby('Brand').size()
print('i. Total numbers of orders placed-')
print()
print(Orders_per_Brand)
print()
Retail_Price_per_Brand= ECom_Data.groupby('Brand')['retail_price'].sum()
print('ii. Total Retail Price per Brand-')
print()
print(Retail_Price_per_Brand)
print()
Discounted_Price_per_Brand= ECom_Data.groupby('Brand')['discounted_price'].sum()
print('iii. Total Discounted Price per Brand-')
print()
print(Discounted_Price_per_Brand)
print()
Brand_Revenue_per_Brand=ECom_Data.groupby('Brand')['Brand_Revenue'].sum()
print('iv. Total Brand Revenue per Brand-')
print()
print(Brand_Revenue_per_Brand)
Brand_DF= pd.DataFrame(columns=['Orders per Brand','Retail Price per Brand','Discounted Price per Brand','Brand Revenue per Brand'])
Brand_DF['Orders per Brand']=Orders_per_Brand
Brand_DF['Retail Price per Brand']= Retail_Price_per_Brand
Brand_DF['Discounted Price per Brand']=Discounted_Price_per_Brand
Brand_DF['Brand Revenue per Brand']=Brand_Revenue_per_Brand
sns.pairplot(data=Brand_DF, vars=['Orders per Brand','Retail Price per Brand','Discounted Price per Brand','Brand Revenue per Brand'], corner=True);
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
ECom_Data['Order_Date'] = pd.to_datetime(ECom_Data['Order_Date'], dayfirst=True)
ECom_Data['Order_Date'].dt.strftime('%d-%m-%Y')
ECom_Data['Months']= ECom_Data['Order_Date'].dt.month
plt.title('Lineplot')
plt.figure(figsize=(20,7))
plt.xticks(np.arange(0,13,1))
sns.lineplot(data=ECom_Data, x='Months',y='Company_Revenue',hue='Region', errorbar=('ci', False))
plt.xlim(1,13);
```
